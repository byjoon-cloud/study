#### apply
- 함수를 사용해서 함수의 리턴값으로 Series를 만들어 새로운 컬럼을 생성 및 수정

```python
def get_domain(email):
    return email.split("@")[1].split(".")[0]

get_domain("andy@gmail.com")
```

```python
df["domain"] = df["email"].apply(get_domain)
df
```

```python
df.drop(columns=["domain"], inplace = True)
df
```

```python
df["domain"] = df["email"].apply(
    lambda email: email.split("@"),[1].split("."),[0])

df
```

```python
# Merge
- user_df, moeny_df 을 만들기
- user_df, money_df를 merge
- merge된 df 를 groupby("money").agg("sum")
```

```python
user_df = pd.DataFrame(columns=["UserID","Name","Age"])

for idx in range(1,6):
    name = md.get_name()
    
    while name in list(user_df["Name"]):
        name = md.get_name()
        
    data = {"Name": name, "UserID": idx, "Age": md.get_age()}
    #data = [idx, name, md.get_age()]
    user_df.loc[len(user_df)] = data

user_df
```

```python
money_df = pd.DataFrame(columns=["ID","Money"])

for idx in range(10):
    money = np.random.randint(1,21) * 1000
    money_df.loc[len(money_df)] = [np.random.randint(1, 6), money]

print(money_df["ID"].unique())
money_df
```

```python
money_df.merge(user_df, left_on="ID", right_on="UserID")
```

```python
user_df.rename(columns={"UserID":"ID"}, inplace=True)
```

```python
user_df.merge(money_df)
```

```python
result_df = user_df.merge(money_df)
result_df
```

```python
result_df.groupby("Name").agg("sum").reset_index()[["ID","Name","Money"]]
```

```python
#outer - merge
```

```python
result_df = pd.merge(user_df,money_df,how="outer")
result_df = result_df.fillna(value=0)
result_df = result_df.groupby("Name").agg("sum").reset_index()[["Name","Money"]]
result_df
```
