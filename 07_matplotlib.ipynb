```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns #인터페이스를 향상
sns.set()
```

```python
iris = sns.load_dataset("iris")
iris.tail(2)
```

```python
sns.lmplot("sepal_length",'sepal_width', data=iris, hue="species", fit_reg =False)
plt.show()
```

```python
sns.lmplot("petal_length",'petal_width', data=iris, hue="species", fit_reg = False)
plt.show()
```
